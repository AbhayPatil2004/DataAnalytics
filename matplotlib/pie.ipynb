```python
import matplotlib.pyplot as plt 

import matplotlib.pyplot as plt

brands = ["samsung", "iphone", "popo", "cococ"]
pop = [25, 40, 78, 65]
colors = ["r", "g", "b", "y"]
ex = [0,0,0,0.1]

plt.pie(pop, labels=brands, colors=colors, autopct='%1.1f%%', startangle=90 , explode = ex , shadow=True )
plt.title("Market Share by Brand")
plt.axis('equal')  # Ensures pie is drawn as a circle
plt.show()
```
