```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

# Create line chart
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Sales')

# Add labels and title
plt.xlabel("Time (Days)")
plt.ylabel("Sales (in $1000)")
plt.title("Sales Over Time")
plt.legend()

# Show the chart
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Months of the year
months = [
    "Jan", "Feb", "Mar", "Apr", "May", "Jun",
    "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
]

# Sample monthly revenue (in $1000)
revenue = [50, 55, 60, 65, 58, 75, 80, 85, 78, 90, 95, 100]

# Create the line chart
plt.figure(figsize=(10, 6))
plt.plot(months, revenue, marker='o', linestyle='-', color='green', label='Revenue')

# Add labels and title
plt.title("Monthly Revenue in 2025", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Revenue (in $1000)", fontsize=12)
plt.grid(True)
plt.legend()

# Highlight max and min revenue
max_value = max(revenue)
min_value = min(revenue)
max_month = months[revenue.index(max_value)]
min_month = months[revenue.index(min_value)]

plt.annotate(f'Max: {max_value}k', xy=(max_month, max_value), xytext=(max_month, max_value + 5),
             arrowprops=dict(facecolor='green', shrink=0.05))
plt.annotate(f'Min: {min_value}k', xy=(min_month, min_value), xytext=(min_month, min_value - 10),
             arrowprops=dict(facecolor='red', shrink=0.05))

# Save the chart as an image
plt.savefig("monthly_revenue_chart.png", dpi=300)

# Show the chart
plt.show()
```
