```python
import matplotlib.pyplot as plt
import numpy as np

labels = ['A', 'B', 'C']
boys = [20, 34, 30]
girls = [25, 32, 34]

x = np.arange(len(labels))  # label locations
width = 0.35                # bar width

plt.bar(x - width/2, boys, width, label='Boys')
plt.bar(x + width/2, girls, width, label='Girls')

# plt.xlabel('Group')
# plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(x, labels)
plt.legend()
plt.show()
```
