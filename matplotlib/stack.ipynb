```python
import matplotlib.pyplot as plt 

days = [ 1 , 2 , 3 , 4 , 5 , 6 , 7 ]
n1 = [ 5 , 10 , 20 , 40 , 50 , 87 , 5 ]
n2 = [ 45 , 1 , 30 , 20 , 80 , 97 , 4 ]
n3 = [ 45 , 5 , 30 , 27 , 80 , 77 , 4 ]

plt.stackplot( days , n1 , n2 , n3 , baseline='sym'  , labels=["week1" , "week2" , "week3"])
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
labels = ['A', 'B', 'C', 'D']
math_scores = [20, 34, 30, 35]
science_scores = [25, 32, 34, 20]
english_scores = [10, 15, 20, 25]

# Plot
x = range(len(labels))

plt.bar(x, math_scores, label='Math')
plt.bar(x, science_scores, bottom=math_scores, label='Science')
bottom2 = [i + j for i, j in zip(math_scores, science_scores)]
plt.bar(x, english_scores, bottom=bottom2, label='English')

# Labels and Title
plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Stacked Bar Chart Example")
plt.xticks(x, labels)
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
days = [1, 2, 3, 4, 5]
sleeping = [7, 8, 6, 11, 7]
eating = [2, 2, 3, 2, 2]
working = [7, 8, 7, 8, 7]
playing = [8, 6, 8, 3, 8]

# Labels for the legend
labels = ['Sleeping', 'Eating', 'Working', 'Playing']

# Create stack plot
plt.stackplot(days, sleeping, eating, working, playing, labels=labels)

# Adding title and labels
plt.title('Daily Activities')
plt.xlabel('Day')
plt.ylabel('Hours')

# Show legend
plt.legend(loc='upper left')

# Show plot
plt.show()
```
