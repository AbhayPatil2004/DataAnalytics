```python
import matplotlib.pyplot as plt 

days = [ 1 , 2 , 3 , 4 , 5 , 6 , 7 ]
n1 = [ 5 , 10 , 20 , 40 , 50 , 87 , 5 ]
n2 = [ 45 , 1 , 30 , 20 , 80 , 97 , 4 ]
n3 = [ 45 , 5 , 30 , 27 , 80 , 77 , 4 ]

plt.stackplot( days , n1 , n2 , n3 , baseline='sym' , color = ["red" , "green" , "blue"] , labels=["week1" , "week2" , "week3"])
plt.legend()
plt.show()
```
