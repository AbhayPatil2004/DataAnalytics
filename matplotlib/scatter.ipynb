```python
import matplotlib.pyplot as plt 
import numpy as np

x = np.random.randint(1, 10, 50)
y = np.random.randint(1, 100, 50)
colors = np.random.randint(10, 100, 50)
size = np.random.randint(10, 100, 50)

plt.scatter(x, y, marker="*", c=colors, cmap="hot" , s = size)  
plt.colorbar()  # display color bar
plt.title("Colored Scatter Plot with Star Marker")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.grid(True)
plt.show()
```
